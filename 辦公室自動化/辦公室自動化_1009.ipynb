### sqlite3 資料庫操作

```python
import sqlite3

db = sqlite3.connect("practice2024.db")
cursor = db.cursor()#創建連結

#創建資料表company，設定欄位名與欄位type
cursor.execute('''
CREATE TABLE COMPANY
       (Store_ID    CHAR(50),
       Salesperson  CHAR(50),
       Sales         INT
       );

''')
db.commit()
```

### 刪除資料表

```python
cursor.execute('''
DROP TABLE COMPANY
''')
db.commit()
```
